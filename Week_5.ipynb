### **Fixing Training Data Size**

```python
training_data_len = math.ceil(len(dataset)*.8)
training_data_len
```

### **Data Normalization**

**Using MinMax Scaler to scale the data between 0 & 1**

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
print(scaled_data.shape)
print(scaled_data)
```

### **Splitting the Dataset**

```python
train_data = scaled_data[0:training_data_len, :]
print(len(train_data))
x_train, y_train = [], []

for i in range(100, len(train_data)):
  x_train.append(train_data[i-100:i, 0])
  y_train.append(train_data[i,0])
  if i<=101:
    print(x_train)
    print(y_train)
```

```python
test_data = scaled_data[training_data_len: , :]
print(len(test_data))
x_test, y_test = [], []

for i in range(100, len(test_data)):
  x_test.append(test_data[i-100:i, 0])
  y_test.append(train_data[i,0])
  if i<=101:
    print(x_test)
    print(y_test)
```

### **Shaping Training & Testing Data**

```python
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1], 1))
print(x_train.shape, y_train.shape)
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
print(x_test.shape, y_test.shape)
```

### **RNN Model (LSTM)**

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape = (x_train.shape[1], 1)))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
```

**Compiling the Model**

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

**Training the Model**

```python
lstm_model = model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size = 64, epochs=100, verbose=1)
```

**Visualizing Loss**

```python
plt.plot(lstm_model.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

**Predicting Values**

```python
train_predictions = model.predict(x_train)
test_predictions = model.predict(x_test)
```

```python
train_predictions = scaler.inverse_transform(train_predictions)
test_predictions = scaler.inverse_transform(test_predictions)
```

```python
print(train_predictions.shape, test_predictions.shape)
```

### **Evaluation Metrics**

```python
math.sqrt(mean_squared_error(y_train,train_predictions))
```

```python
math.sqrt(mean_squared_error(y_test,test_predictions))
```

### **Visualising the Predicted Values**

```python
# train predictions
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[100:len(train_predictions)+100, :] = train_predictions
# test predictions
testPredictPlot = np.empty_like(dataset)
testPredictPlot[:] = np.nan
testPredictPlot[len(train_predictions)+(100*2):len(dataset), :] = test_predictions

plt.figure(figsize = (19,9))
plt.title('RNN model LSTM')
plt.ylabel('Close Price in $', fontsize = 20)

plt.plot(dataset)
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.legend(['Data', 'Train', 'Predictions'], loc='lower right')
plt.show()
```
